```python
rows = [
    [1, 45], [2, 50], [3, 55], [4, 60], [5, 65],
    [6, 70], [7, 75], [8, 80], [9, 85], [10, 90],
    [11, 92], [12, 94], [13, 96], [14, 97], [15, 98],
    [16, 99], [17, 100], [18, 100], [19, 100], [20, 100]
]
cols = ['hours revised', 'exam score']
import pandas as pd
data = pd.DataFrame(data = rows, columns = cols)
data
```

```python
data['exam score'].plot(figsize = (3, 3))
```

```python
data['hours revised'].plot(figsize = (3, 3))
```

```python
import numpy as np

X = data['hours revised']
y = data['exam score']

X_mean = np.mean(X)
y_mean = np.mean(y)

numerator = np.sum((X - X_mean) * (y - y_mean))
denominator = np.sum((X - X_mean) ** 2)
m = round(numerator / denominator, 3)
b = round(y_mean - (m * X_mean), 3)

y_pred = m * X + b

mse = np.mean((y - y_pred) ** 2)
rmse = round(np.sqrt(mse), 3)
mae = round(np.mean(np.abs(y - y_pred)), 3)

ss_res = np.sum((y - y_pred) ** 2)
ss_tot = np.sum((y - y_mean) ** 2)
r_squared = round(1 - (ss_res / ss_tot), 3)

for i in y:
    error = y - y_pred
    gradient_m = -(2 / len(y)) * np.sum(error * X)
    m -= round(0.01 * gradient_m)

for i in y:
    error = y - y_pred
    intercept = -(2 / len(y)) * np.sum(error)
    b -= round(0.01 * intercept)

m, b
r_squared, mae, rmse, round(mse, 3)
```

```python
learning_rate = 0.01
epochs = 100

for _ in range(epochs):
    y_pred = m * X + b
    error = y - y_pred
    gradient_m = -(2 / len(y)) * np.sum(error * X)
    gradient_b = -(2 / len(y)) * np.sum(error)

    m -= learning_rate * gradient_m
    b -= learning_rate * gradient_b

m, b
```
